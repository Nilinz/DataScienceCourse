```python
import pandas as pd

url="https://drive.usercontent.google.com/download?id=1JMYqXipZpz9Y5-vyxvLEO2Y1sRBxqu-U&export=download&authuser=0"
df = pd.read_csv(url)
```

```python
#Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)
df.isnull().sum()
```

```python
print(df.columns)
```

```python
#Видаліть усі стовпці з пропусками, крім стовпця "Мова програмування"
cols_to_drop = df.columns[df.isnull().any()].difference(['Язык.программирования'])
df = df.drop(cols_to_drop, axis=1)
df.head(3)
```

```python
#Знову порахуйте, яка частка пропусків міститься в кожній колонці і переконайтеся, що залишився тільки стовпець "Мова.програмування"
df.isnull().sum()
```

```python
df
```

```python
#Видаліть усі рядки у вихідній таблиці за допомогою методу dropna
df.dropna(inplace=True)
df.isnull().sum()
```

```python
#Визначте новий розмір таблиці за допомогою методу shape

df.shape
```

```python
#Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python
python_data = df[df['Язык.программирования'].str.lower() == "python"]
python_data.head()
```

```python
#Визначте розмір таблиці python_data за допомогою методу shape
python_data.shape
```

```python
#Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"
posada = python_data.groupby(['Должность'])
posada.head(5)
```

```python
#Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"
posada['Зарплата.в.месяц'].agg(['min', 'mean', 'max']).sort_values("mean", ascending=False).apply(lambda x: pd.to_numeric(x, errors='coerce').astype(int) )
```

```python
#Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"
def fill_avg_salary(group):
    avg_salary = group['Зарплата.в.месяц'].mean()
    return pd.Series(avg_salary, index=group.index)
df['avg'] = df.groupby('Должность').apply(fill_avg_salary).reset_index(level=0, drop=True)
df.head(3)
```

```python
#Створіть описову статистику за допомогою методу describe для нового стовпчика.
print(df['avg'].describe())
```

```python
#Збережіть отриману таблицю в CSV файл
df.to_csv('Hw2.2.csv', index=False)
```

Візуалізація

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 8))
position_counts = df['Должность'].value_counts()
plt.pie(position_counts, labels=position_counts.index, autopct='%1.1f%%', startangle=140,  labeldistance=1.2, pctdistance=0.8)
plt.title('Розподіл за посадами')
plt.show()
```

```python
average_salary_by_position = df.groupby('Должность')['Зарплата.в.месяц'].mean()
plt.figure(figsize=(10, 6))
average_salary_by_position.plot(kind='bar', color='skyblue')
plt.title('Середня зарплата за посадою')
plt.xlabel('Посада')
plt.ylabel('Середня зарплата в місяць')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
average_salary_by_age = df.groupby('Возраст')['Зарплата.в.месяц'].mean()
plt.figure(figsize=(10, 6))
average_salary_by_age.plot(kind='bar', color='gray')
plt.title('Середня зарплата за віком')
plt.xlabel('Вік')
plt.ylabel('Середня Зарплата.в.месяц')
plt.xticks(rotation=0)
plt.show()
```
